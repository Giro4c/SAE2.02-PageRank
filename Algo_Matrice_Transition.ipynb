## Version sans commentaire
Attention ! Il faut que la matrice donnée soit constituée de nombres décimaux :  
- Si vous avez un 1 remplacez par 1.0  
- Si vous avez un 0 remplacez par 0.0  

Au sinon la fonction reverra une matrice nulle remplie de 0.  
  
Pour plus d'explication, regarder la partie testing.

```python
from numpy import *

def nombreLinks(M, j):
    count = 0
    for i in range(len(M)):
        count = count + M[j][i]
    return count;

def transitionMatrice(M, a):
    P = M
    N = len(P)
    for i in range(N):
        for j in range(N):
            if (nombreLinks(M, j) != 0):
                P[i][j] = a*M[i][j] + ((1 - a)/N)
            else:
                P[i][j] = 1/N
    return P

def pageRankMatrice(M):
    Q = M
    N = len(P)
    for i in range(N):
        for j in range(N):
            if (nombreLinks(M, j) != 0):
                Q[i][j] = Q[i][j] / nombreLinks(M, j)
            else:
                Q[i][j] = 0
    return Q
```
